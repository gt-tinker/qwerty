## Classical Functions

````qwerty
f: cfunc = lambda x: x
````

## Pipeline Syntax
```
g(f(x))     <==>     x | f | g
         Equivalent
```
Thus, these two expressions are equivalent:

````qwerty
'0'+'1' | reduce | measure
````

````qwerty
measure(reduce('0'+'1'))
````

## Deutsch's Algorithm
Suppose $f(x) = x$. Expected answer: $f(0) \oplus f(1) = 0 \oplus 1 = 1$.

````qwerty
f: cfunc = lambda x: x
````

````qwerty
'0'+'1' | f.sign | reduce | measure
````

## Analyzing Constant Zero Black Box

Suppose $f(x) = 0$. Expected answer: $f(0) \oplus f(1) = 0 \oplus 0 = 0$.

````qwerty
f: cfunc = lambda x: bit(0)
````

Running full algorithm:

````qwerty
'0'+'1' | f.sign | reduce | measure
````

Running first two steps:

````qwerty
'0'+'1' | f.sign
````
