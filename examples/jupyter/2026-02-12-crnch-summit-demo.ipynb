## Deutsch's Algorithm

Suppose $f(x) = x$. Expected answer: $f(0) \oplus f(1) = 0 \oplus 1 = 1$.

```qwerty
f: cfunc = lambda x: x
```

```qwerty
'0'+'1' | f.sign | flop | measure
```

### How does it work?

```qwerty
'0'+'1' | f.sign
```

```qwerty
('0') + (-'1') | flop
```

### Other black boxes

Suppose $f(x) = 0$. Expected answer: $f(0) \oplus f(1) = 0 \oplus 0 = 0$.

```qwerty
f: cfunc = lambda x: bit[1](0b0)
```

```qwerty
'0'+'1' | f.sign | flop | measure
```

Suppose $f(x) = \mathrm{NOT} \: x$. Expected answer: $f(0) \oplus f(1) = 1 \oplus 0 = 1$.

```qwerty
f: cfunc = lambda x: ~x
```

```qwerty
'0'+'1' | f.sign | flop | measure
```

Suppose $f(x) = 1$. Expected answer: $f(0) \oplus f(1) = 1 \oplus 1 = 0$.

```qwerty
f: cfunc = lambda x: bit[1](0b1)
```

```qwerty
'0'+'1' | f.sign | flop | measure
```
